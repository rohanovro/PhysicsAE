# Notebook 1: Data Exploration

Explore the Paderborn bearing dataset — visualise FFT spectra and class distributions.

```python
import sys
sys.path.insert(0, '../src')

import numpy as np
import matplotlib.pyplot as plt
from config import CONFIG, FAULT_FREQUENCIES
from data_loader import load_paderborn_data

ZIP_PATH = '../data/paderborn-db.zip'
```

```python
X_train, X_val, X_fault, scaler = load_paderborn_data(ZIP_PATH, CONFIG)
freq_axis = np.fft.rfftfreq(CONFIG['n_fft'], d=1.0/CONFIG['fs_target'])
```

```python
# Mean FFT spectra: healthy vs fault
fig, ax = plt.subplots(figsize=(12, 4))
ax.plot(freq_axis, X_train.mean(axis=0), label='Healthy', alpha=0.8)
for group, Xf in X_fault.items():
    ax.plot(freq_axis, Xf.mean(axis=0), label=f'{group} fault', alpha=0.6)
for name, ff in FAULT_FREQUENCIES.items():
    ax.axvline(ff, linestyle='--', alpha=0.3)
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('FFT Magnitude')
ax.set_title('Mean FFT Spectra: Healthy vs Fault')
ax.legend(fontsize=8)
plt.tight_layout()
plt.savefig('../results/figures/mean_spectra.png', dpi=150)
plt.show()
```

```python
# Class sizes
print(f'Train (healthy): {X_train.shape[0]:,}')
print(f'Val   (healthy): {X_val.shape[0]:,}')
for g, Xf in X_fault.items():
    print(f'Fault {g}: {Xf.shape[0]:,}')
```
