# Notebook 4: SHAP Explainability Figures

Generate frequency-aware SHAP plots for the paper.

```python
import sys
sys.path.insert(0, '../src')

from config import CONFIG, FAULT_FREQUENCIES
from data_loader import load_paderborn_data
from models import PhysicsAE
from shap_analysis import compute_shap_values, plot_mean_shap
from utils import load_model

ZIP_PATH = '../data/paderborn-db.zip'
```

```python
X_train, X_val, X_fault, scaler = load_paderborn_data(ZIP_PATH, CONFIG)

device = CONFIG['device']
model = load_model(PhysicsAE(input_dim=513, latent_dim=32), '../results/models/physicsae.pt', device)
```

```python
# SHAP on inner fault samples
import numpy as np
X_inner = X_fault['inner'][:100]   # first 100 fault windows

shap_vals = compute_shap_values(
    model, X_train, X_inner, CONFIG,
    n_background=200, seed=42
)
```

```python
plot_mean_shap(
    shap_vals,
    fs=CONFIG['fs_target'],
    n_fft=CONFIG['n_fft'],
    fault_freqs=FAULT_FREQUENCIES,
    title='Mean |SHAP| – Inner Race Fault',
    save_path='../results/figures/shap_inner_fault.png'
)
```
